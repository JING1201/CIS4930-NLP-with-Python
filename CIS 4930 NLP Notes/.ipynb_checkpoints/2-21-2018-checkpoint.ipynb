```python
import nltk
from nltk.corpus import brown, cmudict
```

```python
pronun_entries = nltk.corpus.cmudict.entries()
cmu_words = [i[0] for i in pronun_entries]
news = [w for w in brown.words(categories = 'news') if w.isalpha()]
def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
possible_words = intersection(cmu_words, news)
```

```python
def getRhymingDictionaries (words):
    dict1 = {}
    dict2 = {}
    for word, pron in pronun_entries:
        if word in words:
            dict1[word] = ''.join(pron[1:])
            if dict1[word] in dict2.keys():
                dict2[dict1[word]].append(word)
            else:
                dict2[dict1[word]] = [word]
    return dict1, dict2
```

```python
dict1, dict2 = getRhymingDictionaries(possible_words)
```

```python
for v in dict2.values():
        if len(v) < 4 :
            for w in v:
                if w in possible_words:
                    possible_words.remove(w)
```

```python
news = [w for w in brown.words(categories = 'news') if w.isalpha()]
reverse_news = reversed(news)
bigrams = list(nltk.bigrams(reverse_news))
cfd = nltk.ConditionalFreqDist(bigrams)
```

```python
from nltk.corpus import stopwords
import random
stopwords = stopwords.words("english")
start_word = possible_words[random.randint(0, len(possible_words)-1)]
while start_word in stopwords:
    start_word = possible_words[random.randint(0, len(possible_words)-1)]
```

```python
def generateModel (cfd, word, num):
    temp = word
    sequence = [word]
    while len(sequence)<num:
        sequence = [word]
        for i in range (1, num):
            possibles = cfd[temp].most_common(int(len(cfd)*0.2))
            #print(possibles)
            if len(possibles) == 0:
                break
            else:
                sequence.append(possibles[random.randint(0,len(possibles)-1)][0])
                temp = sequence[-1]
    return list(reversed(sequence))
```

```python
generateModel(cfd, start_word, 7)
```

```python
second_word = dict2[dict1[start_word]][random.randint(0, len(dict2[dict1[start_word]])-1)]
while second_word == start_word:
    second_word = dict2[dict1[start_word]][random.randint(0, len(dict2[dict1[start_word]])-1)]
second_word
```

```python
generateModel(cfd, second_word, 7)
```

```python
third_word = possible_words[random.randint(0, len(possible_words)-1)]
while third_word in stopwords or dict1[third_word]==dict1[start_word]:
    third_word = possible_words[random.randint(0, len(possible_words)-1)]
third_word
```

```python
generateModel(cfd, third_word, 7)
```

```python
fourth_word = dict2[dict1[third_word]][random.randint(0, len(dict2[dict1[third_word]])-1)]
while fourth_word == third_word:
    fourth_word = dict2[dict1[third_word]][random.randint(0, len(dict2[dict1[third_word]])-1)]
fourth_word
```
