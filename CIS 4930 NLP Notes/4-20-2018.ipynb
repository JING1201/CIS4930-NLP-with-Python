# Context Free Grammar
- terminal symbols: { +, -, \*, /, (, ), Z}
- non-terminal symbols: {

- The expressions on the left side can be replaced by the right side
    - exp -> Z
    - exp -> (exp)
    - exp -> exp + exp
    - exp -> exp * exp
    - exp -> exp - exp
    - exp -> exp / exp
    - <--loop--> -> for (<--statement-->; <--condition-->; <--statement-->){
    - <--statement--> -> <--var--> = Z
    - examples:
        - exp = (4+2)\*5
        - exp -> exp \* exp
        - exp -> (exp + exp) \* exp

```python
import nltk
```

```python
sent = ['I', 'shot', 'an', 'elephant', 'in', 'my', 'pajamas']
```

```python
groucho_grammar = nltk.CFG.fromstring("""
    S -> NP VP
    PP -> P NP
    NP -> Det N | Det N PP | 'I'
    VP -> V NP | VP PP
    Det -> 'an' | 'my'
    N -> 'elephant' | 'pajamas'
    V -> 'shot'
    P -> 'in'
    """)
parser  = nltk.ChartParser(groucho_grammar)
trees = parser.parse(sent)

for tree in trees:
    print (tree)
```

```python
parser = nltk.RecursiveDescentParser(groucho_grammar)
trees = parser.parse(sent)

for tree in trees:
    print (tree)
```

```python
def traverse(t):
    try:
        t.label()
    except AttributeError:
        print(t, end=" ")
    else:
        #t node is defined for sure
        print ('(', t.label(), end=" ")
        for child in t:
            traverse(child)
        print (')', end= " ")
```

```python
t = nltk.Tree.fromstring('(S (NP Alice) (VP chased (NP the rabbit)))')
```

```python
traverse(t)
```
