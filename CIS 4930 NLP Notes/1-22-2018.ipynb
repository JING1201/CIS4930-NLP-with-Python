```python
cfd = nltk.ConditionalFreqDist( (genre, word) 
                        for genre in brown.categories()
                        for word in brown.words(categories = genre))
```

    Conditional Freq Dist is used to investigate the frequency/usage of words under different contexts (genres in this case)

```python
genres = ['news','reviews','humor']
```

```python
contexts = ['what','when','where','who','why']
```

```python
cfd.tabulate (conditions = genres, samples = contexts)
```

```python
contexts = ['Monday','Tuesday','Wednesday','Thursday','Friday']
```

```python
cfd.tabulate (conditions = genres, samples = contexts)
```
