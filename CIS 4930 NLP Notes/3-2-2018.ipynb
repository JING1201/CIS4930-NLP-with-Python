# Word distance (mean and variance)
- 'We knocked on Donaldson's door' -5
- 'She knocked on the door' -3
- 'They knocked on his door' -3 
- 'He knocked on the metal front door' -5
- boundaries at subject and direct object
    - word distance is counting the tokens in between the subject and the direct object (ex. we and door)
- N = 4
- mean (mu) = 1/4 (3+3+5+5) = 4.0
- Variance 
    - s^2 = Σ (i=1 to n) [(d_i - mu)^2/(n-1)]
    - n: # of observations/ co-occurences words
    - d = offset of occurence i
    - mu = average offset
    - Example above:
        s^2 = ((3-4)^2 + (3-4)^2 + (5-4)^2 +(5-4)^2)/(4-1)
    - low variance shows that what we are seeing tends to be the case
    - https://ncalculators.com/statistics/coefficient-of-variance-calculator.htm
- t-test
    - t = (x_bar - mu)/ sqrt(s^2/N)
    - http://www.statisticshowto.com/one-sample-t-test/
    - H0: What should be true if w1 s2 do not form a collocation (can reverse depends on how we calculate)
    - Compute the probability p that a word pair event would occur if H0 were true
    - accept H0 if p is high, reject H0 is p is low
    - p < 0.05
    - the lower p indicates that w1 w2 clustering is not just a random event
        - indicates that H0 is false
        - that we are more likely to have identified a collocation
    - P(w1, w2) = P(w1) * P(w2) # assume they are independent variables
        - P(w) = # of observations of w / # of words in corpus
    - x_bar = sample mean from real/observed data
    - s^2 = sample variance
    - N = sample size
    - mu = mean of the distribution (expected)
    - Recap: corpus is a sequence of N bigrams
    - samples 1 when bigram of interest occurs
    -         0 when it does not
    - STEPS:
        - 1) Determine the expected mean
        - 2) Measure the observed mean
        - 3) Run the test ... t-test
    - Example:
        - w1 = new; w2 = companies
        - New York Times:
            - new:           15828 times
            - companies:     4675 times
            - new companies: 8 times
            - Bigram Total:  14307668 times
        - P(new) = 15828/14307668
        - P(companies) = 4675/14307668
        - s^2 to P => P(w1, w2) #approximation
        - P(new, companies) = P(new) * P(companies) =  mu (expected mean)
        - x_bar = (8)/14307668
        - t = (x_bar - mu)/ sqrt(s^2/N)
        - Result: t is high - > accept H0 that these two words are not a collocation
     - t = (x1_bar - x2_bar)/sqrt((s1^2/n1) + (s2^2/n2))
         - Practice:
             - w1 = strong; w2 = powerful; w = force
                 - combine w1 with w, w2 with w and compare
