Sentence Raw -> Segmentation -> (List of Strings) -> Tokenization -> List of list of strings -> POS Tagging -> List of tuples - > Entity Recognition -> Chunked Sentences (Tuples) -> Relations Recognition

```python
import nltk
```

```python
import re
```

```python
def process (text):
    sentences = nltk.sent_tokenize(text)
    sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
    sentences = [nltk.pos_tag(sentence) for sentence in sentences]
    return sentences
```

```python
expression = "NP: {<DT>?<JJ>*<NN>}"
chunker = nltk.RegexpParser(expression)
```

```python
import os
fid = open('AngelasAshes.txt','r')
data = fid.read()
```

```python
sentences = process(data)
sentences
```

```python
sentences[0]
```

```python
results = [chunker.parse(sentence) for sentence in sentences]
results[0]
```

```python
results[0].draw()
```

```python
grammar = r"""
    NP: {<DT|PP\$>?<JJ>*<NN>}
        {<NNP>+}
        """
```

```python
chunker = nltk.RegexpParser(grammar)
results = [chunker.parse(sentence) for sentence in sentences]
results[0]
```

```python
results[0].draw()
```

```python
grammar = r"""
    NP: {<DT|PP\$>?<JJ>*<NN>}
        {<NNP>+}
    VP: {<TO><VB>}
    """
```

```python
chunker = nltk.RegexpParser(grammar)
results = [chunker.parse(sentence) for sentence in sentences]
results[0]
```

```python
results[2].draw()
```

## IOB Tagging
- B - beginning of a phrase (chunk)
- I - inside of a phrase (chunk)
- O - outside of a phrase (chunks)
- ex. George (B-NNP) Washington (I-NNP) Carver (I-NNP)

## Context Free Grammar (CFG)
- Production Rules
    - x -> y
    - start symbol
    - non-terminal symbol
    - terminal symbol

```python
grammar1 = nltk.CFG.fromstring("""
  S -> NP VP
  VP -> V NP | V NP PP
  PP -> P NP
  V -> "saw" | "ate" | "walked"
  NP -> "John" | "Mary" | "Bob" | Det N | Det N PP
  Det -> "a" | "an" | "the" | "my"
  N -> "man" | "dog" | "cat" | "telescope" | "park"
  P -> "in" | "on" | "by" | "with"
  """)
```
