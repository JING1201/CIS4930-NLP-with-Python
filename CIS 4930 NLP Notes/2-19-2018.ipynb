## Problem 1

```python
import nltk
from nltk.corpus import gutenberg
text = [w for w in gutenberg.words('austen-emma.txt') if w.isalpha()]
for i in (0, len(text)-1):
    text[i] = text[i].lower()
```

```python
bigrams = nltk.bigrams(text)
```

```python
bigrams_freq = nltk.FreqDist(bigrams)
bigrams_freq
```

```python
indiv_freq = nltk.FreqDist(text)
indiv_freq
```

```python
relative_freq = {}
for key in bigrams_freq.keys():
    relative_freq[key] = bigrams_freq[key]/(indiv_freq[key[0]]+indiv_freq[key[1]])
```

```python
max(relative_freq, key=relative_freq.get)
```

## Solution to Problem 1

```python
def getMostProbableBigram(words):
    
    bigramList = list(nltk.bigrams(words))
    fdist1 = nltk.FreqDist(words)
    fdist2 = nltk.FreqDist(bigramList)
    
    maxRatio = 0
    maxBigram = None
    ratios = []
    
    for bigram in bigramList:
        ratioNum = float(fdist2[bigram])
        ratioDen = fdist1[bigram[0]] + fdist1[bigram[1]]
        ratio = ratioNum/ratioDen
        ratios.append(ratio)
        
        if ratio > maxRatio:
            maxRatio = ratio
            maxBigram = bigram
    for i in range(10):
        print(bigramList[i], ' ', ratios[i])
   
    return maxBigram
```

```python
getMostProbableBigram(text)
```

```python
from nltk.corpus import brown
news = [w for w in brown.words(categories = 'news') if w.isalpha()]
```

```python
reverse_news = reversed(news)
bigrams = list(nltk.bigrams(reverse_news))
```

```python
cfd = nltk.ConditionalFreqDist(bigrams)
```
